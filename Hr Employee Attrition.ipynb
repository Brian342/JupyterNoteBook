# HR Employee Attrition
- <b><h4>Definition</h4></b> HR analytics is the process of collecting and analyzing Human Resource (HR) data in order to improve an organization’s workforce performance.
The process can also be referred to as talent analytics, people analytics, or even workforce analytics. This method of data analysis takes data that is routinely collected by HR and correlates it to HR and organizational objectives. Doing so provides measured evidence of how HR initiatives are contributing to the organization’s goals and strategies.

- <b><h4>Need of HR Analytics</h4></b> Most organizations already have data that is routinely collected, so why the need for a specialized form of analytics? Can HR not simply look at the data they already have? Unfortunately, raw data on its own cannot actually provide any useful insight. It would be like looking at a large spreadsheet full of numbers and words. Once organized, compared and analyzed, this raw data provides useful insight.They can help answer
- <i><b>questions like:</i></b>

1) What patterns can be revealed in employee turnover?
2) How long does it take to hire employees?
3) What amount of investment is needed to get employees up to a fully productive speed?
4) Which of our employees are most likely to leave within the year?
5) Are learning and development initiatives having an impact on employee performance?

- <b><h4>The process of HR Analytics</b></h4>

HR Analytics is made up of several components that feed into each other.

1) To gain the problem-solving insights that HR Analytics promises, data must first be collected.
2) The data then needs to be monitored and measured against other data, such as historical information, norms or averages.
3) This helps identify trends or patterns. It is at this point that the results can be analyzed at the analytical stage.
4) The final step is to apply insight to organizational decisions.

```python
# 1) Distribution of Atrrition by Age
class kdesubplot:
    def __init__(self, data):
        self.data = data

    def createkdeplot(self):
        fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6,5))

        self.kdeploting(axes, 'Age')

        plt.title('Distrubution of Age by Attrition')
        plt.tight_layout()
        plt.show()

    def kdeploting(self, ax, Col_Age):
        sns.kdeplot(
            data=self.data,
            x = self.data[Col_Age],
            hue='Attrition',
            fill=True,
            ax=ax
        
        )
if __name__ == "__main__":
    kdeplot = kdesubplot(df_copy)
    kdeplot.createkdeplot()
```

```python
# boxplot for salary hikes
sns.boxplot(data=df_copy, x='PercentSalaryHike', y='Attrition', hue='Attrition')
plt.title('Impact of Attrition on salary Hike')
plt.xlabel('PercentSalaryHike (%)', fontsize=10)
plt.ylabel('Attrition', fontsize=10)
plt.show()
```

```python
# correlation between the promotion, SalaryHike, and attrition
plt.subplots(figsize=(8, 6))
df_copy.loc[:,'Attrition_Number'] = df_copy['Attrition'].apply(lambda x: 1 if x == 'Yes' else 0)
sns.heatmap(data=df_copy[['Attrition_Number', 'YearsSinceLastPromotion', 'PercentSalaryHike']].corr(), annot=True, cmap='coolwarm')
plt.title('Numerical Correlation between AttritionNum, Promotion and SalaryHike')
plt.show()
```

```python
# using a chi-square test to determine if there is a significant association between the frequency of promotions and attrition

from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_copy['YearsSinceLastPromotion'], df_copy['Attrition'])

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-Square Statistic: {chi2}, p-value: {p}")
# interpretation: A low p-value(<0.05) indicates a significant association, between promotions and attritions
```

```python
# use a t-test to compare the mean salary hikes of employees who left and those who stayed
from scipy.stats import ttest_ind

# Split data into two groups
salary_hike_left = df_copy[df_copy['Attrition'] == 'Yes']['PercentSalaryHike']
salary_hike_stayed = df_copy[df_copy['Attrition'] == 'No']['PercentSalaryHike']

# Perform t-test
t_stat, p_value = ttest_ind(salary_hike_left, salary_hike_stayed)
print(f"T-Statistic: {t_stat}, p-value: {p_value}")
# A low p-value(<0.05) indicates a significant difference in salary hikes between employees who left and who stayed
```
